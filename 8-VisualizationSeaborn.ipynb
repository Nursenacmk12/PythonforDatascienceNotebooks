```python
import seaborn as sns
data = pd.read_csv('athlete_events.csv.zip')
```

```python
sns.set_style("whitegrid")
sns.scatterplot( x = "Height" , y="Weight" , data=data)
plt.xlabel("height of athletes")
plt.ylabel("height of athletes")
plt.title("Athletes weight vs height")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot( x = "Height" , y="Weight" , data=data)
plt.xlabel("height of athletes")
plt.ylabel("height of athletes")
plt.title("Athletes weight vs height")
plt.show()
```

```python
sns.set_style("whitegrid")  # arka plan rengi
sns.scatterplot( x = "Height" , y="Weight", hue ="Sex", data=data)  # vurgulamak istedigimiz şeyi belirtmek için hue
plt.xlabel("height of athletes")
plt.ylabel("height of athletes")
plt.title("Athletes weight vs height")
plt.show()
```

```python
sns.set_style("whitegrid")  # arka plan rengi
sns.scatterplot( x = "Height" , y="Weight", hue ="Medal", data=data)  # vurgulamak istedigimiz şeyi belirtmek için hue
plt.xlabel("height of athletes")
plt.ylabel("height of athletes")
plt.title("Athletes weight vs height")
plt.show()
```

```python
sns.set_style("whitegrid")  # arka plan rengi
sns.scatterplot( x = "Height" , y="Weight", hue ="Sex", style = "Medal" , data=data)  # vurgulamak istedigimiz şeyi belirtmek için hue
plt.xlabel("height of athletes")
plt.ylabel("height of athletes")
plt.title("Athletes weight vs height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot( x = "Height" , y="Weight" , hue= "Sex", data=data)
plt.xlabel("height of athletes")
plt.ylabel("height of athletes")
plt.title("Athletes weight vs height")
plt.show()
```

```python
sns.set_style("dark")
sns.displot(x = "Height" , hue= "Sex", data=data)
plt.ylabel("Frequency")
plt.title("Athletes height Histogram")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x = "Height" , hue= "Sex", data=data , kind = "kde")
plt.ylabel("Frequency")
plt.title("Athletes height Histogram")
plt.show()
```

```python
sns.barplot( x ="Medal", y="Height", hue="Sex", data=data)
plt.title("medals by height")
plt.show()
```
