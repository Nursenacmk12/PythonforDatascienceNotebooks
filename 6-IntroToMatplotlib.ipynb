```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [10,20,30,30,30,70,40,50,60]
```

```python
weight_list = [30,40,55,60,88,80,89,99,97]
```

```python
plt.plot(age_list , weight_list , "g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("age and weight")
plt.savefig("test.png")
```

```python
numpy_arr1 = np.linspace(0,10,20)
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1 **3
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1 , numpy_arr2)
plt.show()
```

```python
plt.plot(numpy_arr1 , numpy_arr2,"y*-" )
plt.show()
```

```python
plt.plot(numpy_arr2 , numpy_arr1,"y*-" )
plt.show()
```

```python
plt.subplot(1,2,1)  #bir figür içinde birden fazla grafik (subplot) çizmek için kullanılır.
 #plt.subplot(nrows, ncols, index)
plt.plot(numpy_arr1 , numpy_arr2,"r*-" )
plt.subplot(1,2,2)
plt.plot(numpy_arr2 , numpy_arr1,"g--" )
plt.show()
```

```python
my_figure= plt.figure()   # figure oluşturuldu 
figure_axes = my_figure.add_axes([0.1 ,0.1,0.3,0.3])   # buradaki rakamlar genişlik
figure_axes.plot(numpy_arr1 , numpy_arr2,"g")
figure_axes.set_xlabel("X axis")
figure_axes.set_ylabel("Y axis")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_fig = plt.figure( dpi =100)  # çözünürlük
new_axes = new_fig.add_axes([0.2,0.2,0.7,0.7])  
new_axes.plot(numpy_arr1, numpy_arr1**2, label ="numpy_arr2**2")
new_axes.plot(numpy_arr1, numpy_arr1**3, label ="numpy_arr2**3")
new_axes.legend()  # küçük bilgilendirme yeri oluşturur
new_axes.set_xlabel("X axis")
new_axes.set_ylabel("Y axis")
new_axes.set_title("Graph Title")
plt.show()
```
