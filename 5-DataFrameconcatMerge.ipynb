```python
def salary_category(salary):
    if salary < 50000:
        return 'Low'
    elif 50000 < salary < 80000:
        return 'medium'
    else:
        return'high'
        
```

```python
df3["Salary"].apply(salary_category)
```

```python
df3["salary_category"] = df3["Salary"].apply(salary_category)
```

```python
df3
```

```python
# if experience > 10   -> performance_score +1  
```

```python
def adjust_performance(experience):
    if experience > 10:
        return 1
    else:
        return 0
```

```python
 df3["Experience"].apply(adjust_performance)
```

```python
df3["Adjusted"] = df3["Experience"].apply(adjust_performance)
```

```python
df3
```

```python
df3["New_score"] = df3["Performance_Score"] + df3["Adjusted"]
```

```python
df3
```

```python
def adjusted_new(row):
    if row["Experience"] > 10:
        return row["Performance_Score"] + 1
    else:
        return row ["Performance_Score"]
        
```

```python
df3["Adjusted_score"] = df3.apply(adjusted_new , axis = 1)
```

```python
df3
```
