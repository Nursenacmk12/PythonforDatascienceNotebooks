```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Box Plot
# median , quartile , min max 1.5 *IQR outlier
```

```python
data = np.array([5,7,9,11,15,25,29,32,45,55,60,60,60,60,60,60,60,60,61,62])
plt.figure(figsize = (6,5))
sns.boxplot(y = data)
plt.title("Box Plot")
plt.ylabel("data value")
#plt.grid(True)
plt.show()
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
plt.figure(figsize = (13,5))

plt.subplot(1,2,1)
sns.boxplot( x = "class",  y = "age", data = df)
plt.title("age by class")
plt.xlabel("Class")
plt.ylabel("Age")

plt.subplot(1,2,2)
sns.boxplot( x = "class" , y = "fare", data = df)
plt.title("fare by class")
plt.xlabel("Class")
plt.ylabel("Fare")
plt.show()
```
